**Задание 5. В двух столбцах есть пропущенные значения. Один из них — `days_employed`. Пропуски в этом столбце вы обработаете на следующем этапе. Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`. Например, у человека с типом занятости `сотрудник` пропуск в столбце `total_income` должен быть заполнен медианным доходом среди всех записей с тем же типом.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### Категоризация данных

**Задание 16. На основании диапазонов, указанных ниже, создайте в датафрейме `data` столбец `total_income_category` с категориями:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


**Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`. Используйте собственную функцию с именем `categorize_income()` и метод `apply()`.**

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

**Задание 18. Создайте функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

**Например, если в столбце `purpose` находится подстрока `'на покупку автомобиля'`, то в столбце `purpose_category` должна появиться строка `'операции с автомобилем'`.**

**Используйте собственную функцию с именем `categorize_purpose()` и метод `apply()`. Изучите данные в столбце `purpose` и определите, какие подстроки помогут вам правильно определить категорию.**

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

### Шаг 3. Исследуйте данные и ответьте на вопросы

#### 3.1 Есть ли зависимость между количеством детей и возвратом кредита в срок?

```python
func = data.pivot_table(index = ['children'], values = 'debt', aggfunc = {'count','sum'})
func = func.rename(columns={'count':'total','sum':'func'})
func['share'] = round(func['func']/func['total'], 3)
func = func.sort_values(by='children', ascending=True)
print(func)
```

**Вывод:** Прослеживается небольшая зависимость между возратом кредита в срок и количеством детей. Люди без детей показывают наименьший результат по просрочке ввиду своего независимого финансового положения.У людей с 1-2 детьми приблизительно одинаковые просрочки, на их примере можно подтвердить гипотезу. Семьи с 3 или 5 детьми либо имеют меньший процент задержки, либо не берут кредиты вообще. У людей с 4 детьми самые высокие показатели, скорее всего из-за финансового положения.

#### 3.2 Есть ли зависимость между семейным положением и возвратом кредита в срок?

```python
func = data.pivot_table(index = ['family_status'], values = 'debt', aggfunc = {'count','sum'})
func = func.rename(columns={'count':'total','sum':'func'})
func['share'] = round(func['func']/func['total'], 3)
func = func.sort_values(by='share', ascending=True)
print(func)
```

**Вывод:** Первые 3 категории, скорее всего, финансово состоятельные, платежеспособные. Люди, состоящие в гражанском браке или не женатые, имеют вероятность задержки выше, возможно, ввиду малого дохода, молодого возраста.

#### 3.3 Есть ли зависимость между уровнем дохода и возвратом кредита в срок?

```python
func = data.pivot_table(index = ['total_income_category'], values = 'debt', aggfunc = {'count','sum'})
func = func.rename(columns={'count':'total','sum':'func'})
func['share'] = round(func['func']/func['total'], 3)
func = func.sort_values(by='total_income_category', ascending=True)
print(func)
```

**Вывод:** Люди с доходом 30-50 тысяч имеют самую низкую вероятность задержки, возможно, ввиду того, что низкие зарплаты не будут в последствии способны покрыть штраф. Доход в 50-200 тысяч, относительно остальных, имеет большую вероятность невозврата, скорее всего это связано с детьми, количеством кредитов. От 200 тысяч до 1 миллиона люди с хорошим финансовым положением, имеют малую вероятность просрочки. Люди с доходом более миллиона имеют среднюю вероятность не вернуть кредит, что примечательно. Ну и категория с самым низким доходом имеет самую высокую вероятность просрочки.

#### 3.4 Как разные цели кредита влияют на его возврат в срок?

```python
func = data.pivot_table(index = ['purpose_category'], values = 'debt', aggfunc = {'count','sum'})
func = func.rename(columns={'count':'total','sum':'func'})
func['share'] = round(func['func']/func['total'], 3)
func = func.sort_values(by='share', ascending=True)
print(func)
```

**Вывод:** Здесь наблюдается зависимость. У людей, стремящихся к созданию семьи, покупке своей недвижимости, вероятность просрочки ниже, чем у людей, берущих кредит на автомобиль или образование.
